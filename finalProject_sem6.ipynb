```python
# Finding IQR
per25=df['city_pop'].quantile(0.25)
per75=df['city_pop'].quantile(0.75)
iqr=per75-per25
(('per25',per25),
 ( 'per75',per75), 
 ('iqr',iqr))
```

```python
#CAPPING THE OUTLIERS:   
new_df_cap= df.copy()
new_df_cap['city_pop']=np.where(
    new_df_cap['city_pop']>ul,
    ul,
    np.where(
    new_df_cap['city_pop']<ll,
    ll,
    new_df_cap['city_pop']

    )
)
```

Outliers Detection and Removal

```python

# Before IQR
print("Before outlier removal:")
print(df['city_pop'].describe())
print("\nShape of DataFrame before outlier removal:", df.shape)

# Finding IQR
per25 = df['city_pop'].quantile(0.25)
per75 = df['city_pop'].quantile(0.75)
iqr = per75 - per25

# Finding min and max of outlier
ul = per75 + 1.5 * iqr
ll = per25 - 1.5 * iqr

# Finding outliers
outliers = df[(df['city_pop'] > ul) | (df['city_pop'] < ll)]

# Removing outliers and updating the DataFrame
df = df[(df['city_pop'] <= ul) & (df['city_pop'] >= ll)]

# After outlier removal
print("\nAfter outlier removal:")
print(df['city_pop'].describe())

# Alternatively, you can cap the outliers instead of removing them
# df['city_pop'] = np.where(df['city_pop'] > ul, ul, np.where(df['city_pop'] < ll, ll, df['city_pop']))

# You can also check the shape of the DataFrame after outlier removal
print("\nShape of DataFrame after outlier removal:", df.shape)
```

PERFORMING PIPELINE:

```python
# Step 1: train_test_split (which we already did)
# Step 2: OneHotEncoder
from sklearn.compose import ColumnTransformer
trf1=ColumnTransformer([
    ('ohe_category_state',OneHotEncoder(handle_unknown='ignore'),[0,2,4])
],remainder='passthrough'
)
```

```python
# Step 3: Feature Selection
from sklearn.feature_selection import SelectKBest,chi2
trf2=SelectKBest(score_func=chi2,k=5)
```

```python
# Step 4: Train the model using DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
trf3=DecisionTreeClassifier()
```

```python
# Step 5: Now, all chains are created and merge it using pipelines
from sklearn.pipeline import Pipeline
pipe= Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3)
])
```

```python
# Step 6: train & fit 
from sklearn import set_config
set_config(display='diagram')
pipe.fit(X_train,y_train)
```

```python

# Step 7: Predict 
y_pred = pipe.predict(X_test)
```

```python
# Step 8: Calculate accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
# once, we removed the feature selection i.e. step 3 , then the accuracy will be more i.e. 0.9963045846706516
```

```python
# Step 9: GridSearch
params= {'trf3__max_depth':[1,2,3,4,5,None]}
from sklearn.model_selection import GridSearchCV
grid= GridSearchCV(pipe,params,cv=5,scoring='accuracy')
```

```python
#Step 1: using PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=None, svd_solver="arpack")
X_train_trf=pca.fit_transform(X_train)
X_test_trf=pca.transform(X_test)
X_train_trf.shape
```

```python
#Step 2: Applying algorithm
dc=DecisionTreeClassifier()
dc.fit(X_train_trf,y_train)
```

```python
y_pred=dc.predict(X_test_trf)
```

```python
accuracy_score(y_test,y_pred)
```

```python
for i in range(1,4):
    pca=PCA(n_components=None, svd_solver="arpack")
    X_train_trf=pca.fit_transform(X_train)
    X_test_trf=pca.transform(X_test)

    dc=DecisionTreeClassifier()
    dc.fit(X_train_trf,y_train)

    y_pred=dc.predict(X_test_trf)

    print(accuracy_score(y_test,y_pred))
```
